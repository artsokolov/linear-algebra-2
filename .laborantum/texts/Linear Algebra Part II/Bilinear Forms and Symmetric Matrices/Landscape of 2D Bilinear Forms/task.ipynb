```python
Lambda = np.diag([1, 2])
Q = np.array([
    [1/np.sqrt(2), -1/np.sqrt(2)],
    [1/np.sqrt(2), 1/np.sqrt(2)]
])

### YOUR CODE HERE ###

A = Q @ Lambda @ Q.T

XX = np.linspace(-1, 1, 100)
YY = np.linspace(-1, 1, 100)

XX, YY = np.meshgrid(XX, YY)

XY = np.stack([XX, YY])

XY = XY.reshape([2, -1])

Z = []

for index in range(XY.shape[1]):
    Z.append(XY[:, [index]].T @ A @ XY[:, [index]])

Z = np.array(Z)
Z = Z.reshape(XX.shape)

print(Z.shape)

plt.figure(figsize=(10, 10))
plt.contour(XX, YY, Z, 100)
plt.plot([0, Q[0, 0]], [0, Q[0, 1]])
plt.plot([0, Q[1, 0]], [0, Q[1, 1]])

plt.legend([
    '1 eigen vector',
    '2 eigen vector'])
```
